```python
#Scraping Algiers_Province townships
url_city_1 = "https://en.wikipedia.org/wiki/Algiers_Province"
url_city_2 ="https://fr.wikipedia.org/wiki/Casablanca"
html_1=urlopen(url_city_1)
html_2=urlopen(url_city_2)
Soup_Algiers = BeautifulSoup(html_1,'html.parser')
Soup_Casa = BeautifulSoup(html_2,'html.parser')
Soup_Algiers= Soup_Algiers.find('div',attrs={'class' :'div-col'})
Soup_Casa = Soup_Casa.find_all('table',attrs={'class' :'wikitable'})[1]
```

```python
#creating a list of all townships in Algiers City
def Townships_list(soup):
  lis=soup.find_all('li')
  data=[]
  for li in lis[0:]:
      result = li.text.strip().replace(u'\xa0', u'').replace(';','')
      data.append(result)
  return data
```

```python
Algiers_Townships=Townships_list(Soup_Algiers)
Casablanca_Townships=Townships_list(Soup_Casa)
```

```python
#functions to get nearby venues for a location using foursquare api
def getNearbyVenues(names, latitudes, longitudes, radius=5000):
    LIMIT=2000
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT
            )
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Township', 
                  'Township Latitude', 
                  'Township Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
#Algiers venueq[['Venue category]] is the dataframe column containing all categories of venues in Algiers
Algiers_dummies = pd.get_dummies(Algiers_venues[['Venue Category']], prefix="", prefix_sep="")
Algiers_dummies['Township']=Algiers_venues['Township']
Algiers_dummies
```

```python
# move neighborhood column to the first column
fixed_columns = [Algiers_dummies.columns[-1]] + list(Algiers_dummies.columns[:-1])
Algiers_dummies = Algiers_dummies[fixed_columns]
Algiers_dummies
```

```python
Algiers_grouped = Algiers_dummies.groupby('Township').mean().reset_index()
```

```python
#Get insight of the most common venues in every Township
num_top_venues = 5
for hood in Algiers_grouped['Township']:
    print("----"+hood+"----")
    temp = Algiers_grouped[Algiers_grouped['Township'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
"return most common venues in a township"
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Township']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
Townships_venues_sorted = pd.DataFrame(columns=columns)
Townships_venues_sorted['Township'] = Algiers_grouped['Township']

for ind in np.arange(Algiers_grouped.shape[0]):
    Townships_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Algiers_grouped.iloc[ind, :], num_top_venues)

Townships_venues_sorted
```

```python
#create clustering dataframe that contains categories of venues
Algiers_grouped_clustering = Algiers_grouped.drop('Township', 1)
Algiers_grouped_clustering
```

```python
# set number of clusters
kclusters =7
# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=1).fit(Algiers_grouped_clustering)
# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# add clustering labels
Townships_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
Algiers_merged = township
Algiers_merged = Algiers_merged.join(Townships_venues_sorted.set_index('Township'), on='Township')
```

```python
Algiers_merged['Cluster Labels'].value_counts()
```

```python
Algiers_merged['Cluster Labels']=Algiers_merged['Cluster Labels'].fillna(3).astype(int)
```

```python
# create map
map_clusters = folium.Map(location=[City_latitude,City_longitude], zoom_start=10)
# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(Algiers_merged['Latitude'], Algiers_merged['Longitude'], Algiers_merged['Township'], Algiers_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)       
map_clusters
```
